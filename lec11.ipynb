```python
from datascience import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 11.1 Assessing Models

counts = []
for i in np.arange(10000):
    counts.append(np.round(sample_proportions(100, [0.26,0.74]).item(0) * 100))

black = Table().with_columns(
    'Count in a Random Sample', counts
)
black.hist(bins=np.arange(10, 50, 1))
plt.scatter(8, 0, color='red', s=30)

#help(sample_proportions)

# Mendel's Model
distances = []
for i in np.arange(10000):
    distance = abs(sample_proportions(929, [0.75, 0.25]).item(0) * 100 - 75)
    distances.append(distance)
distances_table = Table().with_columns('Distance', distances)
distances_table.hist()
#distances_table.bin()
plt.scatter(705/929*100-75, 0, color='red', s=30)
```
