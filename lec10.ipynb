```python
from datascience import *
import numpy as np
%matplotlib inline
```

```python
# 10.2 Sampling from a Population

united = Table.read_table('united.csv')
united
united.column('Delay').min()
united.column('Delay').max()
united.hist('Delay', bins=np.arange(-20, 300, 10), unit='minute')

united.sample(10).hist('Delay', bins=np.arange(-20, 300, 10), unit='minute')
united.sample(100).hist('Delay', bins=np.arange(-20, 300, 10), unit='minute')
united.sample(1000).hist('Delay', bins=np.arange(-20, 300, 10), unit='minute')
```

```python
# 10.3

np.median(united.column('Delay'))
np.median(united.sample(1000).column('Delay'))

medians = []
for i in np.arange(5000):
    medians.append(np.median(united.sample(1000).column('Delay')))
    
medians
simulated_medians = Table().with_column('Sample Median', medians)
simulated_medians.hist('Sample Median', bins=np.arange(0.5, 5, 1))
```
